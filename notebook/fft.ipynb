Let's plot the FFT of a sine wave!
- choose to use a 1.0s period because nice easy number
- use 64Hz sample frequency because 64 is a smallish + convenient number of samples to transform
- use 4.5Hz frequency because it isn't a divisor of the sample frequency and is therefore much more interesting in the frequency domain
- use 0.2rad phase because it makes the phase plot more interesting

```rust
:dep notebook = { path = "." }
use notebook::*;

let mut signal = 
    BufferedInput::new(synth::sin(SampleRate::new(64), 4.5, 0.2), 64)
    .unwrap();
let fft = dsp::FFTSequence::new(64);
let mut freq: dsp::PolarFFT = fft.fft(&signal.next().unwrap().get_channel(0)).into_polar();
freq.unwrap_phase();
let folded: dsp::FoldedFFT = freq.into_folded();
plot_fft(&folded)
```
