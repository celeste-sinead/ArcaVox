```rust
:dep notebook = { path = "." }
use notebook::*;

let mut signal = 
    BufferedInput::new(synth::sin(SampleRate::new(64), 4.5, 0.), 64)
    .unwrap();
plot_period(&signal.next().unwrap())
```

```rust
let fft = dsp::FFTSequence::new(64);
let freq: Vec<Complex<f32>> = fft.fft(&signal.next().unwrap().get_channel(0));

evcxr_figure((640, 480), |root| {
    root.fill(&WHITE)?;
    let mut chart = ChartBuilder::on(&root)
        .margin(20)
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(0usize..64usize, -32f32..32f32)?;

    chart.configure_mesh().draw()?;

    chart.draw_series(LineSeries::new(freq.iter().map(|y| y.re).enumerate(), &RED))
        .unwrap()
        .label("real")
        .legend(|(x,y)| PathElement::new(vec![(x,y), (x + 20,y)], &RED));
    chart.draw_series(LineSeries::new(freq.iter().map(|y| y.im).enumerate(), &BLUE))
        .unwrap()
        .label("imag")
        .legend(|(x,y)| PathElement::new(vec![(x,y), (x + 20,y)], &BLUE));

    chart.configure_series_labels()
        .position(SeriesLabelPosition::UpperRight)
        .background_style(&WHITE.mix(0.8))
        .border_style(&BLACK)
        .draw()?;
    Ok(())
})
```
