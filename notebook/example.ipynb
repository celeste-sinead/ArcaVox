# Demo a plotters plot

```rust
:dep plotters = { version = "^0.3.6", default-features = false, features = ["evcxr", "all_series", "all_elements"] }
extern crate plotters;
use plotters::prelude::*;
use std::iter::zip;
use std::f32::consts::PI;

let t: Vec<f32> = (0..21).map(|i| i as f32 * 0.1).collect();
let y: Vec<f32> = t.iter().map(|t| (t * PI).sin()).collect();

let figure = evcxr_figure((640, 480), |root| {
    root.fill(&WHITE)?;
    let mut chart = ChartBuilder::on(&root)
        .margin(5)
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(0f32..2.0f32, -1f32..1f32)?;

    chart.configure_mesh().draw()?;

    chart.draw_series(LineSeries::new(zip(t, y), &RED))
        .unwrap()
        .label("y = sin(t*PI)")
        .legend(|(x,y)| PathElement::new(vec![(x,y), (x + 20,y)], &RED));

    chart.configure_series_labels()
        .background_style(&WHITE.mix(0.8))
        .border_style(&BLACK)
        .draw()?;
    Ok(())
});
figure
```
